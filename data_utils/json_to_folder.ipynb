```python
import os
import json
import shutil
import glob
from tqdm import tqdm
```

```python
data_path = '/Users/lamhoangtung/cds_data/final/'
```

```python
def create_folder(ds_type):
    os.makedirs(os.path.join(data_path, 'flowable', ds_type, 'left'), exist_ok=True)
    os.makedirs(os.path.join(data_path, 'flowable', ds_type, 'right'), exist_ok=True)
    os.makedirs(os.path.join(data_path, 'flowable', ds_type, 'center'), exist_ok=True)
```

```python
datasets = json.loads(open(os.path.join(data_path, 'combined.json'),'r').read())
datasets
```

```python
for each_sample in datasets:
    each_sample['rgb_img_path'] = each_sample['rgb_img_path'].replace('//','/')
    each_sample['rgb_img_path'] = each_sample['rgb_img_path'].replace('/home/linus/Desktop/data/', data_path)
    if 'class' not in each_sample.keys():
        datasets.remove(each_sample)
```

```python
create_folder('raw')
```

```python
for each_sample in tqdm(datasets):
    if each_sample['class'] == 0:
        old_path = each_sample['rgb_img_path']
        old_file_name = each_sample['rgb_img_path'].split('/')[-1]
        new_file_name = each_sample['rgb_img_path'].split('/')[-3].replace('.','_') + '_' + old_file_name
        new_path = os.path.join(data_path, 'flowable', 'raw', 'left', new_file_name)
        shutil.copyfile(old_path, new_path)
    if each_sample['class'] == 1:
        old_path = each_sample['rgb_img_path']
        old_file_name = each_sample['rgb_img_path'].split('/')[-1]
        new_file_name = each_sample['rgb_img_path'].split('/')[-3].replace('.','_') + '_' + old_file_name
        new_path = os.path.join(data_path, 'flowable', 'raw', 'center', new_file_name)
        shutil.copyfile(old_path, new_path)
    if each_sample['class'] == 2:
        old_path = each_sample['rgb_img_path']
        old_file_name = each_sample['rgb_img_path'].split('/')[-1]
        new_file_name = each_sample['rgb_img_path'].split('/')[-3].replace('.','_') + '_' + old_file_name
        new_path = os.path.join(data_path, 'flowable', 'raw', 'right', new_file_name)
        shutil.copyfile(old_path, new_path)
```

```python
datasets = json.loads(open(os.path.join(data_path, 'test.json'),'r').read())
create_folder('test')
```

```python
angle_range = [(-60, -5.1), (-5, 5), (5.1, 60)]
range_to_class_id = {}
for index, each_range in enumerate(angle_range):
    range_to_class_id[each_range] = index
range_to_class_id
```

```python
for each_sample in datasets:
    for each_range in angle_range:
        if each_sample['angle'] >= each_range[0] and each_sample['angle'] <= each_range[1]:
            each_sample['class'] = range_to_class_id[each_range]
    each_sample['angle'] = 0 if each_sample['angle'] == -0.0 else each_sample['angle']
    each_sample['rgb_img_path'] = each_sample['rgb_img_path'].replace('//','/')
    each_sample['rgb_img_path'] = each_sample['rgb_img_path'].replace('/home/linus/Desktop/data/', data_path)
```
